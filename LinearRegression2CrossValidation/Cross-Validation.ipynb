# **CROSS VALIDATION**

```
from sklearn.model_selection import KFold
import numpy as np
from sklearn.metrics import accuracy_score
```

```
model_logreg_clf=LogisticRegression()
```

```
# KFold Cross Validation approach
kf = KFold(n_splits=5,shuffle=False)
kf.split(X) 
```

```
accuracy_model=[]
# Iterate over each train-test split
for train_index, test_index in kf.split(X):
    # Split train-test
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    # Train the model
    model = model_logreg_clf.fit(X_train, y_train)
    # Append to accuracy_model the accuracy of the model
    y_pred=model_logreg_clf.predict(X_test)
    accuracy_model.append(accuracy_score(y_test,y_pred , normalize=True)*100)
```

```
# Print the accuracy    
print(accuracy_model)
```

```
len(accuracy_model)
```

```
np.mean(accuracy_model)
```

**KFOLD usage**

```
X1=np.array(np.arange(10))
y1=np.array(2*X1)
fold=1

# KFold Cross Validation approach
kf = KFold(n_splits=5,shuffle=True)
kf.split(X1) 

for train_index, test_index in kf.split(X1):
    # Split train-test
    print("Fold ",fold)
    X_train, X_test = X1[train_index], X1[test_index]
    y_train, y_test = y1[train_index], y1[test_index]
    print('Train: %s, Test: %s' % (X_train, X_test))
    fold=fold+1
```
